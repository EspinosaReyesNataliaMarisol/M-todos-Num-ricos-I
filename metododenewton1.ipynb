```python
from math import * #Esto importa todas las funciones y constantes de la biblioteca math
import numpy as np #se importa la biblioteca numpy para la manipulación de vectores y operaciones matemáticas en Python.
import matplotlib . pyplot as plt #se utiliza para crear gráficos y visualizaciones.
```

```python
# la función original es x^3-x-1=0
def g(x):
    return x**(3)-x-1
```

```python
#recordemos que hay que derivar la función g(x) por lo que su derivada queda de la forma:
def G(x):
    return 3*x**2-1 #derivada de g(x)
```

```python
# gráfica para g(x) y su derivada G(x)
x=np.linspace(0,5,100) #Rango que se utilizará como los valores del eje x para graficar la función g(x) y G(x).
plt.plot(x,g(x))#graficar g(x)
plt.plot(x,G(x), color="purple") #graficar G(X)
plt.title("GRÁFICA DE MÉTODO DE NEWTON PARA g(x)=$x^3 -x- 1$") #imprimir titulo de gráfica
plt.grid(color='pink') #activar cuadricula
plt.xlabel("EJE X") #etiqueta eje x
plt.ylabel("EJE Y")#etiqueta eje y
plt.axhline(0, color="black")#mostrar eje x
plt.axvline(0, color="black")#mostrar eje y
plt.show() #mostrar gráfica
```

```python
# función evaluada en los límites [1,2]
f1=g(1)
f2=g(2)
print(f1,f2)
```

```python
# Ingreso datos de entrada para el método a trabajar
p0=1 #valor inicial de p y el valor mínimo del intervalo
b=2 #valor máximo del intervalo
#guarda valores iniciales del error y del número de iteraciones
tol = 0.01 #float(input("Ingrese el valor de la tolerancia: "))
nmax = 100 #float(input("Ingrese el número máximo de iteraciones: "))
error = 100 #float(input("Ingrese el valor del error: "))
niter = 0 #número inicial de iteraciones
```

```python
# Método de Newton

print("# iter\t\t p\t\t f(p)\t\t  error") #imprimir encabezados de la tabla de iteraciones

#ciclo iterativo
while error > tol and niter < nmax:
    p = p0 - g(p0)/G(p0)  # fórmula de método de Newton p = p0 - g(p0) / G(p0)

    error = abs(p - p0)  # calcular el error como la diferencia absoluta entre el nuevo valor p y el anterior p0

# Imprimir los resultados de cada iteración
    print("{0} \t\t {1:6.6f} \t {2:6.6f} \t {3:6.6f}".format(niter, p0, p, error))

# Actualizamos el valor de P0 y el contador de iteraciones
    p0 = p
    niter += 1
# Imprime el resultado final, mostrando la solución aproximada y el error.
print("La solución aproximada utilizando el método de Newton es p = {0:6.6f} con un error de {1:6.6f}".format(p0, error))


```
