Escribir un programa en Python que implemente el método de Steffensen (modificación del método de Aitken) para resolver ecuaciones no lineales.
Como prueba, resuelva la ecuación f(x) = cos(x) - 2x + 3 = 0 con una precisión de 0.00001.

```python
from math import * #Esto importa todas las funciones y constantes de la biblioteca math
import numpy as np #se importa la biblioteca numpy para la manipulación de vectores y operaciones matemáticas en Python.
import matplotlib . pyplot as plt #se utiliza para crear gráficos y visualizaciones.
```

```python
# De la función original que es f(x)=cos(x)-2x+3=0
def g(x):
    return (3+np.cos(x))/2
```

```python
# gráfica para g(x)
x=np.linspace(-9,10,200) #Rango que se utilizará como los valores del eje x para graficar la función g(x)
plt.plot(x, g(x))#graficar g(x)
plt.title("GRÁFICA DE MÉTODO DE STEFFENSEN PARA g(x)=$(3+cos(x))/2$") #imprimir titulo de gráfica
plt.grid(color='pink') #activar cuadricula
plt.xlabel("EJE X") #etiqueta eje x
plt.ylabel("EJE Y")#etiqueta eje y
plt.axhline(0, color="black")#mostrar eje x
plt.axvline(0, color="black")#mostrar eje y
plt.show() #mostrar gráfica
```

```python
f1=g(1)
f2=g(2)
print(f1, f2)
```

```python
# Ingreso datos de entrada para el método a trabajar
n0=1 #valor inicial de p y el valor mínimo del intervalo
b=2 #valor máximo del intervalo
#guarda valores iniciales del error y del número de iteraciones
tol = 0.00001 #float(input("Ingrese el valor de la tolerancia: "))
nmax = 100 #float(input("Ingrese el número máximo de iteraciones: "))
error = 100 #float(input("Ingrese el valor del error: "))
niter = 0 #número inicial de iteraciones
```

```python
# Método de Steffensen

print("# iter\t\t n0\t\t n1\t\t  n2\t\t n\t\t error") #imprimir encabezados de la tabla de iteraciones

#ciclo iterativo
while error > tol and niter < nmax:
    n1= g(n0) # Calcular p1
    n2 = g(n1) # Calcular p2
    n= n0 - (((n1-n0) ** 2) / (n2 - (2 * n1) + n0)) # Calcular p usando la fórmula de Steffensen
    error = abs(n1 - n0)
    print("{0}\t\t {1:6.6f} \t {2:6.6f} \t {3:6.6f} \t {4:6.6f} \t {5:6.6f}".format(niter, n0, n1, n2, n, error)) # Imprimir los resultados de la iteración actual

    # Actualiza el valor de p0
    n0 = n1
    niter += 1
# # Imprime el resultado final, mostrando la solución aproximada y el error asociado
print("La raíz aproximada de la función es: {0:6.6f} con un error de {1:6.6f}".format(n, error))
```
