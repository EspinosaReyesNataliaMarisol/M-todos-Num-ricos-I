# Gr치fica de $f(x)=x - 3^{-x}$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Funci칩n cuadr치tica.
def f(x):
    return x - 3**(-x)
def g(x):
    return x
def h(x):
    return -3**(-x)
```

```python
f(-5), f(1)
```

```python
# Valores del eje X que toma el gr치fico.
x=np.linspace(-5, 5,100)
```

```python
# Graficar ambas funciones.
plt.figure(figsize=(10,10))
plt.plot(x, f(x))
plt.plot(x, g(x))
plt.plot(x, h(x))
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.savefig("output.png") # Guardar gr치fico como im치gen PNG.
# Mostrarlo.
plt.grid(color="black")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Gr치fica de $f(x)=x-3^{-x}$")
plt.show()
```
