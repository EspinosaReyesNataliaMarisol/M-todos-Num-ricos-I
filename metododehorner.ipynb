```python
from math import *#Esto importa todas las funciones y constantes de la biblioteca math
import numpy as np #se importa la biblioteca numpy para la manipulación de vectores y operaciones matemáticas en Python.
import matplotlib.pyplot as plt #se utiliza para crear gráficos y visualizaciones.
```

```python
def p(x): # el polinomio original
  return 2*x**4 - 3*x**2 + 3*x - 4
```

```python
# Derivamos el polinomio original y queda p´(x) ( es decir P(x))
def P(x):
  return 8*x**3 - 6*x + 3
```

```python
# gráfica para p(x) y su derivada P(x)
x=np.linspace(-10,10,100) #Rango que se utilizará como los valores del eje x para graficar la función p(x) y P(x).
plt.plot(x,p(x))#graficar p(x)
plt.plot(x,P(x), color="purple") #graficar P(X)
plt.title("GRÁFICA DE MÉTODO DE HORNER PARA p(x)=$2x^4 -3x^2+3x- 4$") #imprimir titulo de gráfica
plt.grid(color='pink') #activar cuadricula
plt.xlabel("EJE X") #etiqueta eje x
plt.ylabel("EJE Y")#etiqueta eje y
plt.axhline(0, color="black")#mostrar eje x
plt.axvline(0, color="black")#mostrar eje y
plt.show() #mostrar gráfica
```

```python
# Ingreso datos de entrada para el método a trabajar
x0=-2 #valor inicial de p y el valor mínimo del intervalo
#guarda valores iniciales del error y del número de iteraciones
tol = 0.00001 #float(input("Ingrese el valor de la tolerancia: "))
nmax = 100 #float(input("Ingrese el número máximo de iteraciones: "))
error = 100 #float(input("Ingrese el valor del error: "))
niter = 0 #número inicial de iteraciones
```

```python
# Método de Horner

print("# iter\t xn \t\t f(xn) \t\t f'(xn) \t xn+1 \t\t Error") # Imprimir encabezados de la tabla de iteraciones

# Primer valor de xn, usando p0
xn = x0
f_xn = p(xn) #el polinomio evaluado en x0
f_Xn = P(xn) #derivada evaluado en x0 es decir Q(x0)

# Calcular el primer valor de x
xn_1 = xn - f_xn / f_Xn

# Calcular el error
error = abs(xn_1 - xn)

# Imprimir la primera fila (encabezados)
print("{0}\t {1:6.6f} \t {2:6.6f} \t {3:6.6f} \t {4:6.6f} \t {5:6.6f}".format(1, xn, f_xn, f_Xn, xn_1, error))

# Ciclo iterativo del método de Newton
while error > tol and niter < nmax:
    niter += 1

    # Actualizar el valor de Pn
    xn = xn_1

    # Calcular f(xn) y f'(xn)
    f_xn = p(xn)
    f_Xn = P(xn)

    # Calcular el nuevo valor de x usando la fórmula de Newton
    xn_1 = xn - f_xn / f_Xn

    # Calcular el error
    error = abs(xn_1 - xn)

    # Imprimir los valores de la iteración
    print("{0}\t {1:6.6f} \t {2:6.6f} \t {3:6.6f} \t {4:6.6f} \t {5:6.6f}".format(niter + 1, xn, f_xn, f_Xn, xn_1, error))

    if error <= tol:
        break

# Imprimir el resultado final
print("\nLa solución aproximada utilizando el método de Horner es x = {0:6.6f} con un error de {1:6.6f}".format(xn, error))
```
